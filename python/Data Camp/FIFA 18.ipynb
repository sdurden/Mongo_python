```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
from scipy.cluster.vq import whiten

# Cargamos el CSV a la variable fifa
fifa = pd.read_csv('CompleteDataset.csv',encoding='utf-8',low_memory=False)

# Dado que los valores vienen de tipo string además con los siguientes caracteres '€K' 
# Se procede a reemplazarlos por un caracter vacío de tal forma que quede un número en dos nuevas series dentro del DataFrame
fifa['wage']= fifa['Wage'].str.replace('[€KM]','')
fifa['value']=fifa['Value'].str.replace('[€KM]','')

# Esas nuevas series las pasamos a numero
#pd.to_numeric(fifa['wage'])
#pd.to_numeric(fifa['value'])

#print(pd.to_numeric(fifa['wage'])+0)


# Normalizamos los datos en dos nuevas series por medio de whiten  



fifa['escalado_salario']=whiten(pd.to_numeric(fifa['wage'])+0)
fifa['escalado_valor']=whiten(pd.to_numeric(fifa['value'])+0)


fifa.plot(x='escalado_salario', y='escalado_valor',kind='scatter')
plt.show()







```

Checar la media y la desviación estandar de los valores escalados usando el método `describe()` de Pandas

```python
print(fifa[['escalado_salario', 'escalado_valor']].describe())
```
