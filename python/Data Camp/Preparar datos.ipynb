### ¿Por qué preparar datos para el clustering?

Imagna una situación en donde tengas que establecer un grupo de variables con unidades no comparables (como la dimensión de un producto y su precio, Cm vs $).

Incluso si las variables tienen la misma unidad, pueden ser significativamente diferentes en terminos de sus escalas, por ejemplo el gasto que supone de cereales contra viajes en el año.

* La forma bruta de los datos podría llevar a un sesgo en clustering.
* Clusters pueden ser deasiado dependientes a una variable más que otras.

Esto lo solucionamos con un proceso llamado `Normalización`

Este es un proceso en el cual reescalamos los valores de una varible con respecto a una desviación estandard de los datos. La desviación estandard resultante después de a normalización es 1.

El proceso de normalización es simple y se logra através de dividir un valor por su desviación estandard.

`x_new = x / std_dev(x)`

### Normalización en Python

La librería de normalización está disponible en `Scipy` usando el método `whithen` de la clase `vq`

El método se puede aplicar a una lista simple o a una lista de listas, en el segundo caso, lo que hace la función es obtener la desviación estándard de cada lista y aplicar la formula a su lista correspondiente.



```python
from scipy.cluster.vq import whiten

datos = [5,1,3,3,2,3,3,8,1,2,2,3,5]

datos_escalados = whiten(datos)
print(datos_escalados)
```

Vamos a comparar los datos antes y después de ser normalizados. Para eso utilizaremos la clase pyplot de la librería matplotlib.

Por default `pyplot` grafica líneas. Después habilitamos la leyenda en la gráfica y mostramos la gráfica.

El eje X nos dice el número de puntos en la lista mientras el eje Y nos muestra el valor de la serie original y la serie escalada.

```python
# Importamos la librería para graficar
import seaborn as sns
from matplotlib import pyplot as plt

# Marcamos los datos originales y los escalados

plt.plot(datos, label="originales")
plt.plot(datos_escalados, label ='escaldos')

# Mostramos la leyenda y gráfica
plt.legend()
plt.show()
```
