### Tamaño (Sizes)

Hasta ahora el diagrama de dispersión es solo una nube de puntos azules, indistinguibles uno de otro. ¿No sería mejor si el tamaño de los puntos correspondiera al tamaño de la población?

Para lograr esto, añadimos al parámetro `s=` dentro de scatter el tamaño de los puntos, para esto cargamos el tamaño de la población contenida en la serie `pop`

```python
plt.clf()

import numpy as np
np_pop=np.array(pop)
np_pop=np_pop*7

plt.scatter(gdp_cap, life_exp, s =np_pop)
plt.xscale('log')
plt.xlabel('GDP per Capita [USD]')
plt.ylabel('Esperanza de Vida [Años]')
plt.title('Desarrollo Mundial en 2007')
plt.xticks([1000,10000,100000],['1K','10K','100K'])
plt.show()
```
