### Alto de los jugadores

Como gran fan de baseball, decides llamar a la MLB y pedirles un poco más información sobre la estatura de los jugadores principales.
Ellos te envían datos de más de mil jugadores, los cuales están almacenados en una lista convencional de Python

`height_in`. La altura está expresada en pulgadas. Podrías crear un arreglo `numpy` y convertir las unidades a metros?

Para eso se multiplicará el dato por 0.0254

Además se tendrán que crear 1015 datos aleatorios en la variable `height_in`

```python
import random

height_in=[]
weight_lb=[]

for i in range(0, 1014):
    height_in.append(random.randint(67,83))
    weight_lb.append(random.randint(150,290))
    
#Pasamos el arreglo a tipo Numpy

np_height_in=np.array(height_in)
np_weight_lb=np.array(weight_lb)

#lo imprimimos
print(np_height_in)

# Lo convertimos en metros e imprimimos
np_height_m=np_height_in*0.0254

print(np_height_m)
```

### Calcular el Indice de Masa Corporal de los jugadores

Para lograr esto necesitamos calcular

$$BMI= \frac{weight\, (kg)}{height\,(mts)^2}$$

[Markdown para fórmulas matemáticas](https://csrgxtu.github.io/2015/03/20/Writing-Mathematic-Fomulars-in-Markdown/)

Crear un arreglo en numpy para la lista de pesos en libras, multiplicarlos por 0.453592 para pasar los datos de libras a kg. Almacena el resultado como `np_weight_kg` 

Usa la atura y el peso para calcular el BMI, almacena el resultado en el arreglo `bmi` e imprimelo.

```python
# El vector de pesos ya fue creado np_weight_lb

np_weight_kg=np_weight_lb*0.453592

bmi=np_weight_kg/np_height_m**2


print(bmi)
```

### Buscar con condiciones

Para buscar en las listas normales, se puede utilizar el índice en los corchetes. En numpy específicamente se pueden utilizar valores en arreglos booleanos.

El código que calcula el BMI ya está incluido

- Crea un arreglo boleano numpy, el elemento del arreglo debe ser True si el índice de masa es menor a 21. Nombra al arreglo `light`.

- Imprime el arreglo.
- Imprime los BMI de los jugadores utilizando el arreglo `light`.




```python
light=bmi<21

print(light)

print(bmi[light])
```

## Type en Arreglos Numpy

Si se tiene un arreglo bidimensional o ndimensional en numpy, podemos saber la cardinalidad del arreglo con el método `shape`

`bmi.shape`

Podemos acceder a los miembros del arreglo con los corchetes de forma tradicional `bmi[1][0]` o por medio de una comma `bmi[1,0]` 

Supongamos que necesitamos los datos del segundo y tercer miembro de dos arreglos dentro del arreglo, entonces colocamos dos puntos antes de la comma. `bmi[:,1:3]` esto nos da un arreglo de 2x2 

También se podría seleccionar todo el segundo arreglo de esta forma `bmi[1,:]`

### Arreglo bidimensional 2D con Numpy


```python
baseball = [[180, 78.4],
            [215, 102.7],
            [210, 98.5],
            [188, 75.2]]

np_baseball=np.array(baseball)

print(type(np_baseball))

print(np_baseball.shape)



```

* Carguemos los 1014 registros en un arreglo bidimensional en el que cada renglón contenga el peso y la estatura de los jugadores. (Imprimir)

* Mostremos solo la columna del peso (Imprimir) dentro del arreglo bidimensional

```python
baseball=[]
type(baseball)
for i in range(len(bmi)):
    baseball.append( [height_in[i] , weight_lb[i]])
np_baseball=np.array(baseball)    
print(baseball)
#type(baseball)
```

### Unir dos arreglos en 1 

En el ejemplo anterior vimos como al arreglo `baseball[]` se le añadieron dos series  por medio del método `append` y recorriendolos en un ciclo.

Numpy cuenta con un método que adjunta las series de manera más sencilla, por medio del método **np.column_stack()**

Así, tendríamos el mismo resultado con unas cuantas lineas de código menos.

~~~py
baseball=np.column_stack((height_in,weight_lb))
~~~
