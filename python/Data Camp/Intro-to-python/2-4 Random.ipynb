### Random Float

La aleatoriedad tiene muchos usos en ciencia, arte, estadística, criptografía, juegos, apuestas y otros campos. 

Toda la funcionalidad que necesitamos está contenida en `random` un sub paquete de **Numpy** 


`seed()` : Coloca una semilla aleatoria, así los resultados serán reproducibles entre simulaciones.

`rand()` : si no se especifican argumentos, genera un numer flotante aleatorio entre cero y uno.


```python
import numpy as np

np.random.seed(123)

print(np.random.rand())
```

También se puede usar para números enteros con la función `randint()`. La siguiente llamada de función genera números de 4 a 7, el límite superior no está incluido. 

~~~py
import numpy as np
np.random.randint(4, 8)
~~~

### Caminata Aleatoria


```python

# Initialize random_walk
step = 0
random_walk=[]
random_walk.append(step)

# random_walk=[0]
# Complete the ___
for x in range(99) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = max(step - 1,0)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

¿Qué tan bajo se puede llegar?

Hemos codificado para simular una subida al Empire State de a cuerdo a las tiradas de un dado, sin embargo no se puede llegar a un nivel menos a cero.

Una manera típica de resolver esto es utilizando `max()`. Si pasamos dos argumentos, el mayor es el que se regresa, así, para asugurarnos que una variable nunca bajará de 10, escribimos


~~~py
x = max(10,x-1)
~~~

### Visualizar la caminata

```python
import matplotlib.pyplot as plt

plt.plot(random_walk)
plt.xlabel('Tiradas')
plt.ylabel('Pisos Avanzados')
plt.title('Caminata Aleatoria')
plt.show()
```

### Simular multiples caminatas

Una simple caminata aleatoria es una cosa , pero no te dice su tienes la oportunidad de ganar la pusta.

Para obtener una mejor idea de que tan buenas oportunidades tienes de alcanzar 60 pisos podrías simular repetidamente la caminata (10 veces) y almacenar los resultados.


```python
# Numpy is imported; seed is set

# Initialize all_walks (don't change this line)
all_walks = []

# Simulate random walk 10 times
for i in range(500) :

    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)
```

### Visualizar las caminatas  
`all_walks` es una lista de listas: cada sub-lista representa una sola caminata aleatoria. Si se convirtiera esta lista a un arreglo, se podrían generar interesantes gráficas.

```python
# numpy and matplotlib imported, seed set.

# initialize and populate all_walks
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to Numpy array: np_aw
np_aw=np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()
# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t=np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

Aún queda por saber las probabilidades de alcanzar el piso 60 en el Empire State.

Básicamente lo que se está buscando es el punto final de todas las caminatas aleatorias.Esos puntos tienen una cierta distribución que puede visualizrse en un histograma.

Almacenemos este dato en un arreglo `end` de tipo `Numpy`

```python
print(np_aw_t[-1,:])
x=[]
for i in range(11):
    x.append(i*10)

# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]
        
# Plot histogram of ends, display plot
plt.hist(ends)
#plt.xticks(x)

plt.show()
```
