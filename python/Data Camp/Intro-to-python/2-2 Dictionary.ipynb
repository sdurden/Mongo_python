### Diccionarios

Los diccionarios tienen esta estructura

~~~py
diccionario = {  
               key  : valor,
               key2 : valor,
                 .
                 .
               keyn : valor                 
              }
~~~

Para acceder al valor de un diccionario, basta con especificar la clave del diccionario.

~~~py
print(diccionario[key3]) 
~~~

Para saber cuáles son las claves de un diccionario, utilizamos el método `keys()`


~~~py
# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin', 'norway':'oslo' }
~~~

### Manipulación de Diccionario

Existen una serie de métodos que pueden ser utilizados con los diccionarios

~~~py
clear()                      # Borra los elementos del diccionario
copy()                       # Copia el diccionario (hay que asignarlo a una variable)
fromkeys(keys[], value[])    # Crea un diccionario
get()                        # Obtiene el valor de una clave en específico
items()                      # Regresa una lista con tuplas de claves y valores.
keys()                       # Devuelve una lista con las claves del diccionario
pop(clave)                   # borra un elemento en la clave especificada
popitem()                    # borra el último elemento del diccionario
update(iterable)             # Actualiza el diccionario con los valores específicos
values()                     # Regresa una lista con los valores guardados en el diccionario
~~~

#### Añadir un elmento al diccionario
Si queremos añadir un elemento, solo basta con indicar el diccionario la clave y el valor.

`europe['iceland']='reykjavik'`

#### Borrar y Actualizar

Para borrar un elemento del diccionario lo podemos hacer por medio de
  `del(diccionario['clave'])`  

O por medio del método `pop(clave)` o `popitem()` que borra el último elemento insertado en el diccionario.

Para actualizar el contenido de una clave se puede hacer con el método `update` o hacerlo implicitamente.


~~~py
#Ejemplo

europe.update(germany='berlin')
europe['germany']='berlin'
~~~
